```python
import os                              # File and folder operations
import shutil                          # Moving or copying files
import cv2                             # Image loading, resizing, cropping
import numpy as np                     # Image arrays, math operations
import matplotlib.pyplot as plt        # Visual debugging, comparisons

from skimage.metrics import structural_similarity as ssim  # Image similarity
from PIL import Image                  # Optional fallback for image handling

import torch                           # GPU acceleration, if needed later
from torchvision import transforms     # Preprocessing, augmentations if required
```

```python
from google.colab import files
import os

uploaded = files.upload()  # opens a file picker

# Filter for .jpg files
image_paths = [f for f in uploaded.keys() if f.lower().endswith('.jpg')]

print(f"Uploaded {len(image_paths)} JPG images.")
```

```python
os.makedirs("selected", exist_ok=True)
os.makedirs("discarded", exist_ok=True)

# Initialize
selected_images = []
discarded_images = []

resize_dim = (512, 384)  # Resize for faster SSIM comparison
similarity_threshold = 0.10  # Adjust for sensitivity

# Load the first image as reference
ref_img = cv2.imread(image_paths[0], cv2.IMREAD_GRAYSCALE)
ref_img = cv2.resize(ref_img, resize_dim)
selected_images.append(image_paths[0])
shutil.copy(image_paths[0], os.path.join("selected", image_paths[0]))

# Compare remaining images ===
for filename in image_paths[1:]:
    current_img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
    current_img = cv2.resize(current_img, resize_dim)

    score = ssim(ref_img, current_img)

    if score < similarity_threshold:
        selected_images.append(filename)
        ref_img = current_img  # Update reference image
        shutil.copy(filename, os.path.join("selected", filename))
    else:
        discarded_images.append(filename)
        shutil.copy(filename, os.path.join("discarded", filename))

#Summary Report
print("\n====== SUMMARY ======")
print(f" Total uploaded images: {len(image_paths)}")
print(f" Selected: {len(selected_images)}")
print(f" Discarded : {len(discarded_images)}")

if discarded_images:
    print("\n Discarded image names:")
    for img in discarded_images:
        print(" -", img)
```

```python
import random
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Assuming selected_images and image_dir already defined in your environment
kernel_size = 2
padding = kernel_size // 2
resize_dim = (1024, 768)

# Select 10 random images from selected_images (or all if less than 10)
num_samples = min(10, len(selected_images))
sample_images = random.sample(selected_images, num_samples)

for idx, img_path in enumerate(sample_images):
    full_path = os.path.join(image_dir, img_path)
    img = cv2.imread(full_path)
    if img is None:
        print(f"Warning: Could not read image {full_path}")
        continue

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray = cv2.resize(gray, resize_dim)

    # Pad image
    padded_img = np.pad(gray, pad_width=padding, mode='edge')

    h, w = gray.shape
    score_matrix = np.zeros((h, w))

    # Sliding window mean intensity calculation
    for i in range(padding, h + padding):
        for j in range(padding, w + padding):
            neighborhood = padded_img[i - padding:i + padding + 1, j - padding:j + padding + 1]
            score_matrix[i - padding, j - padding] = np.mean(neighborhood) / 255.0

    # Print image name
    print(f"\nImage {idx + 1}: {img_path}")

    # Downsample score matrix for print readability (e.g., every 48 pixels)
    score_sample = score_matrix[::48, ::48]
    print(np.round(score_sample, 2))

    # Optional: Visualize the grayscale and score matrix
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.imshow(gray, cmap='gray')
    plt.title(f"Grayscale Image: {img_path}")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(score_matrix, cmap='gray')
    plt.title("Sliding Window Color Score")
    plt.axis('off')

    plt.show()

    # ---------------------------
    # Edge noise detection part
    # ---------------------------
    height, width = gray.shape
    vertical_band = int(width * 0.2)
    horizontal_band = int(height * 0.2)

    top_band = gray[:horizontal_band, :]
    bottom_band = gray[-horizontal_band:, :]
    left_band = gray[:, :vertical_band]
    right_band = gray[:, -vertical_band:]

    std_top = np.std(top_band)
    std_bottom = np.std(bottom_band)
    std_left = np.std(left_band)
    std_right = np.std(right_band)

    global_std = np.std(gray)

    threshold_factor = 2.0

    crop_top = std_top > threshold_factor * global_std
    crop_bottom = std_bottom > threshold_factor * global_std
    crop_left = std_left > threshold_factor * global_std
    crop_right = std_right > threshold_factor * global_std

    # Convert grayscale to BGR for drawing colored crop lines
    visualization = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)

    if crop_top:
        visualization = cv2.line(visualization, (0, horizontal_band), (width, horizontal_band), (0, 0, 255), 2)
    if crop_bottom:
        visualization = cv2.line(visualization, (0, height - horizontal_band), (width, height - horizontal_band), (0, 0, 255), 2)
    if crop_left:
        visualization = cv2.line(visualization, (vertical_band, 0), (vertical_band, height), (0, 0, 255), 2)
    if crop_right:
        visualization = cv2.line(visualization, (width - vertical_band, 0), (width - vertical_band, height), (0, 0, 255), 2)

    print(f"Crop flags (top, bottom, left, right): {(crop_top, crop_bottom, crop_left, crop_right)}")
    print(f"Std deviations (top, bottom, left, right): "
          f"({std_top:.2f}, {std_bottom:.2f}, {std_left:.2f}, {std_right:.2f})")
    print(f"Global std deviation: {global_std:.2f}")

    plt.figure(figsize=(8, 8))
    plt.title(f"Potential Crop Lines: {img_path}")
    plt.imshow(cv2.cvtColor(visualization, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```
